```python
def calculate_adr(data, window=4):
    high_low = data['High'] - data['Low']
    adr = high_low.rolling(window=window).mean()
    return adr
```

```python
import yfinance as yf
import pandas as pd

def calculate_rsi(data, window=4):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()

    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rs

# data = yf.download('MMM',period='15d')
# calculate_rsi(data)


# #     "AAPL", "ABBV", "ABT", "ACN", "ADBE", "AIG", "ALL", 
# #                  "AMGN", "AMT", "AMZN", "AXP", "BA", "BAC", "BIIB", 
# #                  "BK", "BKNG", "BLK", "BMY", "C", "CAT", "CHTR", 
# #                  "CL", "CMCSA", "COF", "COP", "COST", "CRM", "CSCO", 
# #                  "CVS", "CVX", "DD", "DHR", "DIS", "DOW", "DUK", 
# #                  "EMR", "EXC", "F", "META", "FDX", "GD", "GE", 
# #                  "GILD", "GM", "GOOG", "GOOGL", "GS", "HD", "HON", 
# #                  "IBM", "INTC", "JNJ", "JPM", "KHC", "KMI", "KO", 
# #                  "LLY", "LMT", "LOW", "MA", "MCD", "MDLZ", "MDT", 
# #                  "MET", "MMM", "MO", "MRK", "MS", "MSFT"]

# # 
# # Initialize a list to store stocks that meet the entry criteria
# stocks_meeting_criteria = []

# for ticker in sp100_tickers:
#     data = yf.download(ticker,period='15d',Interval='1d')  # Fetch data for the last 5 days
#     data['RSI'] = calculate_rsi(data['Close'])
# #     print(data['RSI'])
#     # Check if the last day's RSI is below 15
# #     if data['RSI'].iloc[-1] < 15 and data['Close'].iloc[-1] < 100:
# #         stocks_meeting_criteria.append(ticker)

# print(stocks_meeting_criteria)
```
